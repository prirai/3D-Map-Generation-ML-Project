```python
import numpy as np
from matplotlib import pyplot as plt

with open('3D_Map/vietnam1.xyz') as f:
    tmparr = f.readlines()
rows = int(tmparr[-1].split()[0])
cols = int(tmparr[-1].split()[1])
imgz = np.empty(shape=(rows+1, cols+1))
for entry in tmparr:
    entry = entry.split()
    row = int(entry[0])
    col = int(entry[1])
    val = float(entry[2])
    imgz[row][col] = val
```

```python
k = 8
patches = []
for i in range(k):
    for j in range(k):
        patch = imgz[i*rows//k:(i+1)*rows//k, j*cols//k:(j+1)*cols//k]
        patch = patch.flatten()
        patches.append(patch)
imgzF = imgz.flatten()
```

```python
len(patches)
```

```python
import scipy.stats as stats

fig, axes = plt.subplots(k, k, figsize=(10,8))
for i in range(k):
    for j in range(k):
        stats.probplot(patches[i*k+j], plot=axes[i][j], dist='uniform', sparams=(0, 1))
plt.show()
```

```python
stats.probplot(imgzF, dist='uniform', plot=plt)
plt.show()
```
